```python
Y = data.AQI
```

```python
Y.head()
```

**Performing Data Cleaning by substituting NaN values to mean value**

```python
Ymean = data['AQI'].mean()
Y.fillna(Ymean,inplace=True)
```

```python
Y.isnull().sum()
```

**Selecting predictors**

```python
data_predictors = data.drop(['AQI'],axis=1)
X = data_predictors.select_dtypes(exclude=['object'])
```

```python
X.head()
```

```python
X.info()
```

**Splitting train set and test set**

```python
from sklearn.model_selection import train_test_split
X_train,X_valid,Y_train,Y_valid = train_test_split(X,Y,train_size=0.8, test_size=0.2,
                                                      random_state=0)
```

**Data cleaning of predictors**

```python
from sklearn.impute import SimpleImputer

# Imputation
my_imputer = SimpleImputer()
imputed_X_train = pd.DataFrame(my_imputer.fit_transform(X_train))
imputed_X_valid = pd.DataFrame(my_imputer.transform(X_valid))

# Imputation removed column names; put them back
imputed_X_train.columns = X_train.columns
imputed_X_valid.columns = X_valid.columns
```

```python
imputed_X_train.isnull().sum()
```

```python
Y_train.isnull().sum()
```

```python
Y_train.shape
```

```python
X_valid.shape
```

```python
Y_valid.shape
```

**Applying machine Learning models**

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
```

```python
#model 1
model = RandomForestRegressor(n_estimators = 10, random_state=0)
model.fit(imputed_X_train,Y_train)
```

```python
prediction = model.predict(imputed_X_valid)
```

```python
mae = mean_absolute_error(Y_valid,prediction)
mae
```

```python
from sklearn.metrics import mean_squared_error

rms = mean_squared_error(Y_valid, prediction, squared=False)
rms
```

```python
model.score(imputed_X_valid, Y_valid)
```

```python
#model 2
from sklearn import linear_model
from sklearn.metrics import mean_absolute_error
```

```python
model2 = linear_model.LinearRegression()
model2.fit(imputed_X_train,Y_train)
```

```python
prediction2 = model2.predict(imputed_X_valid)
```

```python
mae = mean_absolute_error(Y_valid,prediction2)
mae
```

```python
rms = mean_squared_error(Y_valid, prediction2, squared=False)
rms
```

```python
model2.score(imputed_X_valid, Y_valid)
```

```python
#model3
from sklearn import svm
from sklearn.svm import SVR
model3 = svm.SVR()
model3.fit(imputed_X_train,Y_train)
```

```python
prediction3 = model3.predict(imputed_X_valid)
```

```python
mae = mean_absolute_error(Y_valid,prediction3)
mae
```

```python
rms = mean_squared_error(Y_valid, prediction3, squared=False)
rms
```
